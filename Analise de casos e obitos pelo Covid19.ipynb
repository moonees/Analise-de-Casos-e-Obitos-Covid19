```python
import pandas as pd
low_memory=False
import matplotlib.pyplot as plt
# Mostrar os gráficos no notebook
%matplotlib inline
pd.options.display.max_columns = 100
pd.options.display.max_rows = 100
```

```python
# Achar os dados
arquivo = r'C:\Users\moone\Downloads\HIST_PAINEL_COVIDBR_06jul2020.xlsx'
# Ler os dados
df = pd.read_excel(arquivo)
```

```python
# Agrupando e criando o gráfico
df.groupby('estado')['obitosAcumulado'].max().sort_values().tail(8).plot(kind='bar',figsize=(10,5))
# Alterar titulo do gráfico
plt.title('Estados com o maior numero de obitos')
```

```python
df['data'] = df['data'].dt.strftime('%d/%m')
```

```python
Sp = df.query('municipio == "São Paulo"')
Sp.tail(30).plot(kind='bar',figsize=(12,5), x= 'data', y='obitosAcumulado', color = 'purple')
plt.title('Evolução do numero de obitos na cidade de São Paulo nos ultimos 30 dias')
```

```python
sjcCasos = df.query('municipio == "São José dos Campos"')
sjcCasos.tail(40).plot(kind='bar',figsize=(12,5),x='data',fontsize=14,y='casosAcumulado', color = 'red')
plt.title('Evolução no numero de casos em São José dos Campos')
```

```python
sjcObitos= df.query('municipio == "São José dos Campos"')
sjcObitos.tail(40).plot(kind='bar',figsize=(15,5),x='data',y='obitosAcumulado', color = 'blue')
plt.title('Evolução no numero de obitos em São José dos Campos')
```
